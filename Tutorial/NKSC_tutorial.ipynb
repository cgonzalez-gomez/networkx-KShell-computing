# Tutorial 

This guide can help you start working with NetworkX Kshell Computing.

```python
import Kshell as ks
import networkx as nx
```

## Find the nodes with low degree

```python
G=nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(1,5)])
ks.find_nodes(G,2)
```

## Get all the k-cores of a Graph
The k-core of a graph is the maximal subgraph such that every vertex has degree at least k. The k-shell is the set of vertices that are part of the k-core but not part of the (k+1)-core.

```python
kshell_G=ks.k_shell(G)
kshell_G
```

### Access to the main shell 

```python
kshell_G[-1]
```

## Draw a subgraph of the main shell nodes

### General

```python
ks.draw_kshell(G,kshell_G[-1])
```

### Gene regulatory network
In the study of a gene regulatory network it can be interesting to have different colors to have a visual representation of the nature of a node (target gene or transcription factor).

```python
DG=nx.DiGraph()
target={2,3,4,5}
TF={1,3,2}
DG.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(1,5)])
kshell_DG=ks.k_shell(DG)
ks.draw_kshell(DG,kshell_DG[-1],target=target,TF=TF)
```
